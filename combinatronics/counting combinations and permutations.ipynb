### Counting permutations and combinations

Count permutations when an event has a specific order.

Count combinations when there is no specific order for an event.

#### Combinations

$ n $ be the number of elements in the sample

$ r $ be the number of elements to choose

Then:

$ {n \choose r} = nCr = \frac{n!}{r!(n - r)!} $

But recall the definition of factorials:

$ n! = n(n-1)(n-2)...(n-r)... $

So for efficiency:

$ {n \choose r} = nCr = \frac{\prod_{i=n-r+1}^n{i}}{r!} $

```octave
# EG
n = 12;
r = 2;
disp(factorial(n)/(factorial(r)*factorial(n-r)))
disp(prod(n-r+1:1:n)/factorial(r))
```

### Combinations with replacement

$ {n+r-1 \choose r} = \frac{(n+r-1)!}{r!((n+r-1) - r)!} $

$ {n+r-1 \choose r} = \frac{(n+r-1)!}{r!(n-1)!} $

But since $ n+r-1 > n-1 $ then this simplifies to:

$ {n \choose r}^R = nC^Rr = \frac{\prod_{i=n}^{n+1-1}{i}}{r!} $

```octave
n = 15;
r = 4;
disp(factorial(n+r-1)/(factorial(r)*factorial(n-1)))
disp(prod(n:1:n+r-1)/(factorial(r)))
```

#### Permutations

Choices with order

$ nPr = {n\choose r}r! = \frac{n!}{(n - r)!} $

Simplifies to:

$ \prod_{i=n-r+1}^n{i} $

```octave
n = 10;
r = 3;
disp(factorial(n)/(factorial(n-r)))
disp(prod(n-r+1:1:n))
```
